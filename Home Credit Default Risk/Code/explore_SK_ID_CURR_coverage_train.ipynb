```python
import pandas as pd
import numpy as np
```

```python
bureau = pd.read_csv('bureau.csv')[['SK_ID_BUREAU','SK_ID_CURR']]
bureau_balance = pd.read_csv('bureau_balance.csv')[['SK_ID_BUREAU']]
application_train = pd.read_csv('application_train.csv')[['SK_ID_CURR']]
```

```python
bureau_merge = bureau_balance.merge(bureau, on='SK_ID_BUREAU',how='inner').drop_duplicates()
```

```python
print(bureau['SK_ID_BUREAU'].nunique())
print(bureau_balance['SK_ID_BUREAU'].nunique())
print(bureau_merge['SK_ID_BUREAU'].nunique())
print(bureau['SK_ID_CURR'].nunique())
print(bureau_merge['SK_ID_CURR'].nunique())
```

```python
bureau_merge__application_train = bureau_merge.merge(application_train, on='SK_ID_CURR', how='inner').drop_duplicates()
print(bureau_merge__application_train['SK_ID_CURR'].nunique())
print(bureau_merge['SK_ID_CURR'].nunique())
print(application_train['SK_ID_CURR'].nunique())
```

```python
bureau__application_train = bureau.merge(application_train, on='SK_ID_CURR', how='inner').drop_duplicates()
print(bureau__application_train['SK_ID_CURR'].nunique())
print(bureau['SK_ID_CURR'].nunique())
print(application_train['SK_ID_CURR'].nunique())
```

```python
POS_CASH_balance = pd.read_csv('POS_CASH_balance.csv')[['SK_ID_CURR']]
POS_CASH_balance__application_train = POS_CASH_balance.merge(application_train, on='SK_ID_CURR', how='inner').drop_duplicates()
print(POS_CASH_balance__application_train['SK_ID_CURR'].nunique())
print(POS_CASH_balance['SK_ID_CURR'].nunique())
print(application_train['SK_ID_CURR'].nunique())
```

```python
credit_card_balance = pd.read_csv('credit_card_balance.csv')[['SK_ID_CURR']]
credit_card_balance__application_train = credit_card_balance.merge(application_train, on='SK_ID_CURR', how='inner').drop_duplicates()
print(credit_card_balance__application_train['SK_ID_CURR'].nunique())
print(credit_card_balance['SK_ID_CURR'].nunique())
print(application_train['SK_ID_CURR'].nunique())
```

```python
previous_application = pd.read_csv('previous_application.csv')[['SK_ID_CURR']]
previous_application__application_train = previous_application.merge(application_train, on='SK_ID_CURR', how='inner').drop_duplicates()
print(previous_application__application_train['SK_ID_CURR'].nunique())
print(previous_application['SK_ID_CURR'].nunique())
print(application_train['SK_ID_CURR'].nunique())
```

```python
installments_payments = pd.read_csv('installments_payments.csv')[['SK_ID_CURR']]
installments_payments__application_train = installments_payments.merge(application_train, on='SK_ID_CURR', how='inner').drop_duplicates()
print(installments_payments__application_train['SK_ID_CURR'].nunique())
print(installments_payments['SK_ID_CURR'].nunique())
print(application_train['SK_ID_CURR'].nunique())
```

```python
n = application_train['SK_ID_CURR'].nunique()
print('Dataset','No. of training SK_ID_CURR contained','Coverage %')
print('application_train.csv',n,'NA')
print('bureau.csv',bureau__application_train['SK_ID_CURR'].nunique(),str(round(bureau__application_train['SK_ID_CURR'].nunique()/n*100))+'%')
print('bureau_balance.csv',bureau_merge__application_train['SK_ID_CURR'].nunique(),str(round(bureau_merge__application_train['SK_ID_CURR'].nunique()/n*100))+'%')
print('POS_CASH_balance.csv',POS_CASH_balance__application_train['SK_ID_CURR'].nunique(),str(round(POS_CASH_balance__application_train['SK_ID_CURR'].nunique()/n*100))+'%')
print('credit_card_balance.csv',credit_card_balance__application_train['SK_ID_CURR'].nunique(),str(round(credit_card_balance__application_train['SK_ID_CURR'].nunique()/n*100))+'%')
print('previous_application.csv',previous_application__application_train['SK_ID_CURR'].nunique(),str(round(previous_application__application_train['SK_ID_CURR'].nunique()/n*100))+'%')
print('installments_payments.csv',installments_payments__application_train['SK_ID_CURR'].nunique(),str(round(installments_payments__application_train['SK_ID_CURR'].nunique()/n*100))+'%')
```
